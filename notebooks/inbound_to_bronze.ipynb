### Lendo os dados da camada Inbound

```scala
%python
# Import required libraries
import pyspark.pandas as ps  # pandas API on Spark

```

```scala
val path = "dbfs:/mnt/dados/inbound/dados_brutos_imoveis.json"
val dados = spark.read.json(path)
```

```scala
display(dados)
```

```scala
val dados_anuncio = dados.drop("imagens","usuario")

display(dados_anuncio)
```

### Criando Coluna de identificacao

```scala
%python
from pyspark.sql.functions import col
```

```scala
import org.apache.spark.sql.functions.col
```

```scala
val df_broze = dados_anuncio.withColumn("id", col("anuncio.id"))
display(df_broze)
```

## Salvando na Camada bronze

```scala
val path = "dbfs:/mnt/dados/bronze/dataset_imoveis"
df_broze.write.format("delta").mode(SaveMode.Overwrite).save(path)
```
