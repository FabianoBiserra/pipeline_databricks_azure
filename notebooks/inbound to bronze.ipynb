### Lendo os dados da camada Inbound

```python
# Import required libraries
import pyspark.pandas as ps  # pandas API on Spark

```

```python
df = ps.read_json("dbfs:/mnt/dados/inbound/dados_brutos_imoveis.json")
```

```python

df = df.drop(["imagens","usuario"], axis=1)

#display(df)
```

```python
df["anuncio"]
```

```python
from pyspark.sql.functions import col
```

```python
df_broze = df.to_spark().withColumn("id", col("anuncio.id"))
display(df_broze)
```

## Salvando na Camada bronze

```python
path = "bdfs:/mnt/dados/bronze/dataset_imoveis"
df_broze.write.format("delta").mode("overwrite").save(path)
```
